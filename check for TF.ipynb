```python
import tensorflow as tf
import numpy as np
import matplotlib as mlt
```

```python
from keras.datasets import cifar10

# Load CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()# Print the shapes of the dataset to check
print(x_train.shape)  # Should be (50000, 32, 32, 3)
print(y_train.shape)  # Should be (50000, 1)
print(x_test.shape)   # Should be (10000, 32, 32, 3)
print(y_test.shape)   # Should be (10000, 1)

```

```python
import matplotlib.pyplot as plt  # Add this import
cifar10_classes = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

# Example training images and their labels
print('Example training images and their labels: ' + str([x for x in y_train[0:10]]))

# Corresponding classes for the labels
print('Corresponding classes for the labels: ' + str([cifar10_classes[int(x)] for x in y_train[0:10]]))

fig ,axarr =plt.subplots(1, 10)
fig.set_size_inches(20, 6)

for i in range(10):
    axarr[i].imshow(x_train[i])  # Display image
    axarr[i].set_title(cifar10_classes[int(y_train[i][0])])  # Set the title as the class
    axarr[i].axis('off')  # Remove axis for better visuals

plt.show()  # Show the plot
```

```python
x_train.shape, y_train.shape,x_test.shape,y_test.shape
```

```python
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
from tensorflow.keras.models import Sequential
from tensorflow import keras
from keras.layers import Dense
from keras.layers import Flatten
```

```python
ann =keras.Sequential()
ann.add(Flatten(input_shape=(32,32,30)))
ann.add(Dense(2048,activation='relu'))
ann.add(Dense(10,activation='softmax'))
```

```python
ann.summary()
```

```python
ann.compile(loss='sparse_categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = ann.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test))
```
